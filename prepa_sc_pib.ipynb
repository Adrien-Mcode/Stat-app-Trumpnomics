```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\adxva\OneDrive\Bureau\ENSAE 2A - S1\STAT APP\Stat-app-Trumpnomics\dfR_complete_netexp.csv",
                       header = 0,
                       sep=',')
df
```

```python
w = pd.read_csv(r"C:\Users\adxva\OneDrive\Bureau\ENSAE 2A - S1\STAT APP\Stat-app-Trumpnomics\pond_r_pib.csv",header = 0,
                       sep=',')
w
```

```python
df_pib = df[['LOCATION', 'TIME', 'PIB']]
df_pivot = df_pib.pivot("TIME", "LOCATION", "PIB")
pred = df_pivot.drop('USA', axis = 1).to_numpy() @ w['w.weight']
pred_pib = pd.DataFrame(pred)
pred_pib
```

```python
df_chomage = df[['LOCATION', 'TIME', 'Chomage']]
df_pivot2 = df_chomage.pivot("TIME", "LOCATION", "Chomage")
predcho = df_pivot2.drop('USA', axis = 1).to_numpy() @ w['w.weight']
pred_cho = pd.DataFrame(predcho)
pred_cho
```

```python
#EMPLOIS
df_emplois = df[['LOCATION', 'TIME', 'Emplois']]
df_pivot_emp = df_emplois.pivot("TIME", "LOCATION", "Emplois")
predemp = df_pivot_emp.drop('USA', axis = 1).to_numpy() @ w['w.weight']
pred_emp = pd.DataFrame(predemp)
pred_emp
```

```python
#Actifs
df_actifs = df[['LOCATION', 'TIME', 'Actifs']]
df_pivot_actifs = df_actifs.pivot("TIME", "LOCATION", "Actifs")
predactifs = df_pivot_actifs.drop('USA', axis = 1).to_numpy() @ w['w.weight']
pred_actifs = pd.DataFrame(predactifs)
pred_actifs
```

```python
graph_pib = pd.concat([pred_pib, pred_cho, pred_emp, pred_actifs], axis = 1)
graph_pib.columns = ['sc_pib', 'sc_cho', 'sc_emp', 'sc_act']
graph_pib
```
